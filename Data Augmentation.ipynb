```python
import os
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import array_to_img, img_to_array, load_img
# define the path to the directory containing the original images
data_dir ="D:\\PlantVillage"

# define the path to the directory where augmented images will be saved
augmented_dir = "D:\\New folder"

# define the number of augmented images to generate for each original image
num_augmentations = 20

# define the image size to use for augmentation
img_size = (256, 256)

# define the data augmentation settings
datagen = ImageDataGenerator(
 rotation_range=40,
 width_shift_range=0.2,
 height_shift_range=0.2,
 shear_range=0.2,
 zoom_range=0.2,
 horizontal_flip=True,
 fill_mode='nearest')

# loop over all the subdirectories in the data directory
for subdir in os.listdir(data_dir):
    subdir_path = os.path.join(data_dir, subdir)
    if os.path.isdir(subdir_path):
        # create a new directory for the augmented images
        augmented_subdir = os.path.join(augmented_dir, subdir)
        os.makedirs(augmented_subdir, exist_ok=True)

        # loop over all the original images in the subdirectory
        for filename in os.listdir(subdir_path):
            if filename.endswith(".jpg") or filename.endswith(".png"):
                # load the image
                img = load_img(os.path.join(subdir_path, filename), target_size=img_size)
                x = img_to_array(img)
                x = x.reshape((1,) + x.shape)

                # generate num_augmentations augmented images from the original image
                i = 0
                for batch in datagen.flow(x, batch_size=1, save_to_dir=augmented_subdir, save_prefix=filename[:-4], save_format="jpg"):
                    i += 1
                    if i >= num_augmentations:
                        break

                print("Generated %d augmented images from %s" % (num_augmentations, filename))
```
